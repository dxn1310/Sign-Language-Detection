```python
#importing the required modules
import pickle
import numpy as np
```

```python
data_dict = pickle.load(open('./data.pickle', 'rb')) #getting the hand landmarks data 
```

```python
print(data_dict.keys())
print(data_dict)
```

```python
#Converting to numpy arrays
data = np.asarray(data_dict['data'])
labels = np.asarray(data_dict['labels'])
```

Model Creation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Spliting into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels)
```

```python
model = RandomForestClassifier() #using Random Forest Classifier
```

```python
model.fit(x_train, y_train) 
```

Testing the model

```python
prediction = model.predict(x_test)

score = accuracy_score(prediction, y_test)
```

```python
print('{}% of samples were classified correctly !'.format(score * 100))
```

Saving the model to use for live hand sign detection using webcam or laptop camera

```python
f = open('model.pkl','wb')
pickle.dump({'model':model},f)
f.close()
```
